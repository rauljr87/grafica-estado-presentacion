#### Subplots se trabaja con axes

```python
#Delegamos en una función todo lo que quiero representar
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0.0, 5.0, 0.1)
x2 = np.arange(0.0, 5.0, 0.2)
```

```python
#Una figura con 2 ejes
plt.figure(1) #figura 1

#subplot: 2 filas y 1 columnas y rellenamos los datos del dibujo 1
plt.subplot(2,1,1) #Relación 2 a 1 (x), el doble de ancho que de alto del grafico #1
plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k')

#subplot: 2 filas y 1 colunma y rellenamos los datos del dibujo 2
plt.subplot(2,1,2) #Relación 2 a 1 (x), el doble de ancho que de alto del gráfico #2
plt.plot(x2, f(x2), 'g--')

plt.show()
```

```python
#Una figura con 2 ejes
plt.figure(1) #figura 1

#subplot: 2 filas y 1 columnas y rellenamos los datos del dibujo 1
plt.subplot(2,2,1) #Relación 2 a 1 (x), el doble de ancho que de alto del grafico #1
plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k')

#subplot: 2 filas y 1 colunma y rellenamos los datos del dibujo 2
plt.subplot(2,2,2) #Relación 2 a 1 (x), el doble de ancho que de alto del gráfico #2
plt.plot(x2, f(x2), 'g--')

plt.show()
```

### Textos en los gráficos

```python
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
#Utilizar density para reemplazar a normed
#desity valor booleano

#Para escribir en latex se debe anteponr la letra r

n, bins, patches = plt.hist(x, 50, facecolor = 'g', alpha = 0.6, density = 1)

plt.xlabel("Coeficiente Intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu, \sigma)$")

plt.text(120, 0.015, r'$\mu = 100, \ \sigma=20$') #texto

plt.axis([20, 180, 0, 0.025])
#Pintando una parrilla en el gráfico
plt.grid(True)


plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 90) #gráfico de 10 pulgadas de ancho por 6 de alto y 90 pixeles por pulgada

plt.subplot(1,1,1)

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)

plt.plot(x, y, linewidth = 2.0)

#annotate, escribir una anotación
#xy, coordenadas donde se quiere tener exactamente la punta de una flecha
#xytext, coordenadas del texto
#arrowprops, definir el parámetro de las propiedades de la flecha
#arrowprops es igual a un diccionario de datos
#facecolor, color de la flecha
#shrink, indica la pronunciación de la flecha

plt.annotate("Maximo Local", xy = (4*np.pi, 1), xytext = (15, 1.5), 
            arrowprops = dict(facecolor = "black", shrink = 0.05))

#Abrir el gráfico, expandiendo el limite de las y de -2 a 2
plt.ylim(-2, 2)

plt.show()
```

##### Cambios de ejes y escalas

```python
mu = 0.5
sd = 0.3

y = mu + sd * np.random.randn(1000)
y = y[(y>0) & (y<1)]

y.sort()

x = np.arange(len(y))
```

```python
from matplotlib.ticker import NullFormatter #Elimina formatos de los gráficos, debemos ubicarnos en el formato
```

```python
plt.figure(figsize = (10, 8))

#Escala lineal
plt.subplot(2,2,1)#dos filas, dos columnas, el primer gráfico
plt.plot(x, y)

plt.yscale('linear')#Escala lineal
plt.xscale('linear')#Escala lineal

plt.title("Escala Lineal")

plt.grid(True)

#Escala logaritmica
plt.subplot(2,2,2)
plt.plot(x, y)

plt.yscale("log") #Indicar la escala en y hace que sea una escala semilogaritmica
#plt.xscale("log")

plt.title("Escala Logaritmica")

plt.grid("True")

#Escala simetrica
plt.subplot(2,2,3)
plt.plot(x, y - y.mean())

plt.yscale("symlog", linthreshy = 0.01)#simetric log, linthreshy = divisiones logaritmicas

plt.title("Escala Log Simétrico")

plt.grid("True")

#Escala logística
plt.subplot(2,2,4)#Dos filas, dos columnas, gráfico 4

plt.plot(x, y)

plt.yscale("logit")#logit, curvas logísticas

plt.title("Escala Logística")

#gca, get current axis - formato acual 
plt.gca().yaxis.set_minor_formatter(NullFormatter())#Eliminamos el formato del gráfico actual en el eje de las y

plt.grid(True)

#Ajustar los gráficos, moverlos ligeramente
#hspace, espacio de altura entre los gráficos
#wspace, espacio de anchura entre los gráficos
plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.25, wspace = 0.25)

plt.show()
```

```python
#Tamaño de la figura
plt.figure(figsize = (10, 8))


#Plots de seno y coseno
plt.plot(x, S, color = 'blue', linewidth = 1.2 , linestyle = '-', label = 'seno')#Se debe colocar una label para usar legend
plt.plot(x, C, color = 'green', linewidth = 1.2, linestyle = '-', label = 'coseno')#Se debe colocar una label para usar legend


#Aplica lo mismo que con el eje de las y.
plt.xlim(x.min()*1.1, x.max()*1.1)
#Abrir el gráfico, expandiendo el limite de las y de -1.2 a 1.2
#Se toma el punto minimo del seno multiplicado por 1.1 para el inicio para obtener un 10% más abajo del valor 
#más pequeño de y, y como final se toma el punto máximo del seno multiplicado por 1.1 para obtener un 10% más por ensima
#del valor mas grande de y.
plt.ylim(S.min()*1.1, S.max()*1.1)#Obteniendo un 10% mas por abajo del gráfico


#Marcas o ticks
#linspace para x espaciar
#Creamos un vector para definir de otra forma las divisiones, por ejemplo con los valores de Pi
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],#Divisiones en multiplos de pi y pi medios
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])#Segundo vector con expresiones de latex
#no olvidar la r para el string crudo
plt.yticks(np.linspace(-1, 1, 3, endpoint = True),
          ['-1', '', '+1'])#linspace de -1 hasta 1 y 3 divisiones, endpoint para que tome el final


#Axes
#gca; Get Current Axis, da acceso a los ejes del dibujo actual
ax = plt.gca()
#Acceso a los bordes del gráfico y tine un array de 4 parámetros
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#Nos ubicamos en el eje de las x
ax.xaxis.set_ticks_position('bottom')
#Movemos el eje de las x hasta la posición 0 del eje de las y
ax.spines['bottom'].set_position(('data', 0))#data son los datos originalmente para el eje x y ele eje y queda subida al 0
#Nos ubicamos en el eje de las y
ax.yaxis.set_ticks_position('left')
#Movemos el eje de las y hasta la posición donde correspondería el 0 de los propios datos.
ax.spines['left'].set_position(('data',0))


#Legend
#Colocar una legenda, a los plots se les debe colocar una etiqueta (label) para poder usar esta propiedad
plt.legend(loc = 'upper left')


#Marcaciones en el gráfico
x0 = 2*np.pi/3
#[punto de inicio en x, punto de fin en x], [punto de incio en y, punto de fin en y]
#Desde x0 - 0 hasta x0 - np.sin(x0), una raya vertical de color azul
plt.plot([x0, x0], [0, np.sin(x0)], color = 'blue', linewidth = 2.5, linestyle = '--')


#Colocando un punto gordote en la intercepción de la raya vertical y el seno
#Solo punto de inicio de x = x0 y solo punto de inicio de y = np.sin(x0)
plt.scatter([x0, ], [np.sin(x0), ], 50, color = 'blue')#tamaño 50


#Documentando el seno
#Anotar un texto
#xy, la anotación estará ubicada en la misma posición del punto gordo
#xycoords, para que las coordenadas coincidan con los datos
#xytext, para subir ligeramente el texto (+20, +40)
#textcoords en offset points, para ajustar la ubicación de xytext en 20 puntos más a la derecha 
#de las x y 40 puntos más arriba de las y (puntos del gráfico en lugar de coordenadas absolutas)
#Arrowprops para colocar una fleca.
#Connectionstyle, para el estilo de la conección con una dimensión 
#de arco 3 y ángulo 0.2 radianes con respecto al eje horizontal, así la flecha toma un efecto de arco con un ángulo 
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$', 
            xy = (x0, np.sin(x0)), xycoords = 'data',
            xytext = (+20, +40), textcoords = 'offset points',
            fontsize = 16,
            arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3, rad=.2'))


#Documentando el coseno
plt.plot([x0, x0], [0, np.cos(x0)], color = 'green', linewidth = 2.5, linestyle = '--')
plt.scatter([x0, ], [np.cos(x0), ], 50, color = 'green')
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$', 
            xy = (x0, np.cos(x0)), xycoords = 'data',
            xytext = (-90, -60), textcoords = 'offset points',
            fontsize = 16,
            arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3, rad=.2'))


#Decorando las etiquetas
#Se debe realizar por medio de un bucle ya que hay muchas etiquetas
#ax.get_xticklabels(), axis creado y las marcas de las x
#ax.get_yticklabels(), axis creado y las marcas de las y
#El bucle recorre las etiquetas del eje de las x y del eje de las y al mismo tiempo utilizando
#el signo '+', el signo '+' significa suma de conjuntos,
#se obtiene el conjunto de todas las etiquetas del eje de las x y se obtiene el conjunto de todas
#las etiquetas del eje de las y, y para cada una de ellas las guarda en la variable label y se aplica un tamaño
#de letra a 16 puntos, y le colocamos una caja bbox.
#edgecolor, None para que no ponga un color en el borde de la caja
#alpha, para una transparencia

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor = 'white', edgecolor = 'None', alpha = 0.6))

    
plt.show()
```
