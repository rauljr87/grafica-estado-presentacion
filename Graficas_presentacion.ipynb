```python
import pandas as pd
import numpy as pd
import matplotlib.pyplot as plt
import os

from matplotlib.backends.backend_pdf import PdfPages #Salvar un plot en PDF
```

### Gr√°fica del estado de los Receptores ###

```python
estado_receptores = ['Mal Estado', 'Estado Regular']
valores = [15, 18]
colores = ['silver', 'yellow']
explode_values = [0, 0.1]

plt.pie(x = valores, labels = estado_receptores, colors = colores, autopct='%1.2f%%', shadow=True, explode = explode_values)

plt.title('Estado de Equipos Receptores')

#plt.legend()

plt.savefig('../Anexos_scrip_graficas/Plots/receptores.jpg', transparent = True)

plt.show()
```

### Gr√°fica del estado de las Antenas ###

```python
estado_antenas = ['Buen Estado','Mal Estado', 'Estado Regular']
valores = [6, 11, 32]
colores = ['green', 'silver', 'coral']
explode_values = [0, 0.1, 0]

plt.pie(x = valores, labels = estado_antenas, colors = colores, autopct='%1.2f%%', shadow=True, explode = explode_values)

plt.title('Estado de las Antenas')

plt.savefig('../Anexos_scrip_graficas/Plots/antenas.jpg', transparent = True)

plt.show()
```
