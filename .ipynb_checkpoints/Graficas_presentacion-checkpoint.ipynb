## Gráfica de tipo PIE ##

### Gráfica del estado de los Receptores ###

```python
estado_receptores = ['Mal Estado', 'Estado Regular']
valores = [15, 18]
colores = ['silver', 'yellow']
explode_values = [0, 0.1]

plt.pie(x = valores, labels = estado_receptores, colors = colores, autopct='%1.2f%%', shadow=True, explode = explode_values)

plt.title('Estado de Equipos Receptores')

#plt.legend()

plt.savefig('../Anexos_scrip_graficas/Plots/receptores.jpg', transparent = True)

plt.show()
```

## Gráfica tipo barras ##

```python
help(plt.bar)
```

### Gráfica del estado de los Receptores ###

```python
# Nombres para las ticks
zonales = ['CZO2', 'CZO3', 'CZO4', 'CZO5', 'CZO6', 'OTG']

# Valores para el eje y
estado_malo = [4, 3, 0, 2, 6, 0]
estado_regular = [5, 3, 1, 6, 0, 3]

# Valores para el eje x
indice = np.arange(len(zonales))

# Creación de las primeras barras
plt.bar(x=indice, height= estado_malo, label='Mal Estado', color='red')

# Creación de las segundas barras
plt.bar(x=indice, height= estado_regular, label='Estado Regular', bottom=estado_malo)

# Etiquetado del eje x
plt.xticks(indice, zonales)
# Etiqueta ubicada en el lado de las x
plt.xlabel('Coordinaciones Zonales y Oficina Técnica')
# Etiqueta ubicada en el lade las y
plt.ylabel('Número de equipos Receptores')
# Título
plt.title('Estado de Equipos Receptores por ODs y OTG')

# Legenda
plt.legend()

# Modificando los ejes
plt.axis([-1, 6, 0, 9]) #Modificación de ejes, eje x (-1 - 6), eje y (0 - 9)

##Parrilla
#plt.grid(True)

plt.savefig('../Anexos_scrip_graficas/Plots/receptores_bar.jpg', transparent = True)

plt.show()
```

### Gráfica del estado de las Antenas ###

```python
# Nombres para las ticks
zonales = ['CZO2', 'CZO3', 'CZO4', 'CZO5', 'CZO6', 'OTG']

# Valores para el eje y
estado_bueno = [6, 0, 0, 0, 0, 0]
estado_malo = [0, 2, 0, 0, 9, 0]
estado_regular = [0, 6, 8, 6, 7, 5]

# Valores para el eje x
indice = np.arange(len(zonales))

# Creación de las primeras barras
plt.bar(x=indice, height= estado_bueno, label='Buen Estado', color='green')

# Creación de las segundas barras
plt.bar(x=indice, height= estado_malo, label='Mal Estado', bottom=estado_bueno, color='red')

# Creación de las terceras barras
plt.bar(x=indice, height= estado_regular, label='Estado Regular', bottom=estado_malo)

# Etiquetado del eje x
plt.xticks(indice, zonales)
# Etiqueta ubicada en el lado de las x
plt.xlabel('Coordinaciones Zonales y Oficina Técnica')
# Etiqueta ubicada en el lade las y
plt.ylabel('Número de Antenas')
# Título
plt.title('Estado de Antenas por ODs y OTG')

# Legenda
plt.legend()

plt.savefig('../Anexos_scrip_graficas/Plots/antenas_bar.jpg', transparent = True)

plt.show()
```

### Gráfica del Estado de los Juegos de Adaptadores y cables en general ###

```python
# Nombres para las ticks
zonales = ['CZO2', 'CZO3', 'CZO4', 'CZO5', 'CZO6', 'OTG']

# Valores para el eje y
estado_bueno = [0, 0, 0, 0, 0, 1]
estado_malo = [2, 2, 0, 0, 0, 0]
estado_regular = [0, 0, 2, 2, 2, 0]

# Valores para el eje x
indice = np.arange(len(zonales))

# Creación de las primeras barras
plt.bar(x=indice, height= estado_bueno, label='Buen Estado', color='cyan')

# Creación de las segundas barras
plt.bar(x=indice, height= estado_malo, label='Mal Estado', bottom=estado_bueno, color='red')

# Creación de las terceras barras
plt.bar(x=indice, height= estado_regular, label= 'Estado Regular', bottom=estado_malo)

# Etiquetado del eje x
plt.xticks(indice, zonales)
# Etiqueta ubicada en el lado de las x
plt.xlabel('Coordinaciones Zonales y Oficina Técnica')
# Etiqueta ubicada en el lade las y
plt.ylabel('Número de juegos de Adaptadore y Cables en general')
# Título
plt.title('Estado de Adaptadores y Cables por ODs y OTG')

# Legenda
plt.legend(loc='best')
#plt.legend(loc=(0.75, 0.75))

# Modificando los ejes
plt.axis([-1, 6, 0, 3]) #Modificación de ejes, eje x (-1 - 6), eje y (0 - 3)


##Parrilla
#plt.grid(True)

plt.savefig('../Anexos_scrip_graficas/Plots/adaptadores_bar.jpg', transparent = True)

plt.show()
```

### Gráfica del Estado de las Estaciones SACER ###

```python
# Nombres para las ticks
estaciones = ['CCN', 'CCR', 'EF', 'ERT', 'EM', 'EMT']

# Valores para el eje y
# Al parecer existe un problema al graficar estos valores...
estado_bueno = [0, 2, 0, 2, 1, 0]
estado_malo = [0, 0, 1, 4, 4, 1]
estado_regular = [1, 8, 1, 14, 1, 1]

# Valores para el eje x
indice = np.arange(len(estaciones))

# Creación de las primeras barras
plt.bar(x=indice, height= estado_bueno, label='Buen Estado', color='green')

# Creación de las segundas barras
plt.bar(x=indice, height= estado_malo, label='Mal Estado', bottom=estado_bueno, color='red')

# Creación de las terceras barras
plt.bar(x=indice, height= estado_regular, label= 'Estado Regular', bottom=estado_malo)

# Etiquetado del eje x
plt.xticks(indice, estaciones)
# Etiqueta ubicada en el lado de las x
plt.xlabel('Estaciones del sistema SACER')
# Etiqueta ubicada en el lade las y
plt.ylabel('Número de Estaciones SACER')
# Título
plt.title('Estado de Estaciones del sistema SACER por ODs y OTG')

# Legenda
plt.legend(loc='best')
#plt.legend(loc=(0.75, 0.75))

# Modificando los ejes
plt.axis([-1, 6, 0, 18]) #Modificación de ejes, eje x (-1 - 6), eje y (0 - 3)


##Parrilla
#plt.grid(True)

plt.savefig('../Anexos_scrip_graficas/Plots/estaciones_bar.jpg', transparent = True)

plt.show()
```
