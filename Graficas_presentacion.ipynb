```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from matplotlib.backends.backend_pdf import PdfPages #Salvar un plot en PDF
```

### Gráfica del estado de los Receptores ###

```python
estado_receptores = ['Mal Estado', 'Estado Regular']
valores = [15, 18]
colores = ['silver', 'yellow']
explode_values = [0, 0.1]

plt.pie(x = valores, labels = estado_receptores, colors = colores, autopct='%1.2f%%', shadow=True, explode = explode_values)

plt.title('Estado de Equipos Receptores')

#plt.legend()

plt.savefig('../Anexos_scrip_graficas/Plots/receptores.jpg', transparent = True)

plt.show()
```

### Gráfica del estado de las Antenas ###

```python
zonales = ['CZO2', 'CZO3', 'CZO4', 'CZO5', 'CZO6', 'OTG']
estado_malo = [4, 3, 0, 2, 6, 0]
estado_regular = [5, 3, 1, 6, 0, 3]

indice = np.arange(len(zonales))

# Creación de las primeras barras
plt.bar(x=indice, height= estado_malo, label='Mal Estado')

# Creación de las segundas barras
plt.bar(x=indice, height= estado_regular, label='Estado Regular', bottom=estado_malo)

# Etiquetado del eje x
plt.xticks(indice, zonales)
# Etiqueta ubicada en el lado de las x
plt.xlabel('Coordinaciones Zonales y Oficina Técnica')
# Etiqueta ubicada en el lade las y
plt.ylabel('Número de equipos Receptores')

plt.title('Estado de Equipos Receptores por ODs y OTG')

# Legenda
plt.legend()

plt.show()
```
